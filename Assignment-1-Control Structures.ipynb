```python
BMI=int(input("Enter your BMI"))
```

```python
if(BMI<18.5):
    print("under weight")
elif(BMI<=24.9):
    print("Normal")
elif(BMI<=29.9):
    print("over weight")
elif(BMI<=34.0):
    print("Obesity Class-1")
elif(BMI<=39.9):
    print("Obesity Class-2")
else:
    print("Extreme Obesity")
    
```
